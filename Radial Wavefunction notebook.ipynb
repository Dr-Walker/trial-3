# Radial Wavefunction Notebook

In this notebook we are going to work through the radial part of the wavefunction. This is going to let us explore topics such as electron sheilding and penetration. 

The radial part of the electrons wavefunction depends on the first three quantum numbers, the principle quantum number n, and the angular wavefunction number, l give us how large the orbital is and its distrbution as we go further away from the nucleus. The third quantum number, ml gives us the direction along the x,y,and z axes that the orbital occupies. 

The simplest orbital is the 1s orbital, where, n=1, l=0 and ml=0. For simplicities sake in the this work book we are going to shorten ml to just m.

In spherical coordinates (shown below), the wavefunction, $ \psi $ can be expressed as a factor of two other functions one $ R $, that only depends on the radius away from the center and the spherical part $ Y $ that depends on the angles $ \theta $ and   $\phi $.

##### eq. 1
$ \psi (r, \theta, \phi) = R(r)Y(\theta, \phi) $  




#### Figure 1: Spherical coordinate system 

Since in this workbook we are only conserned with the radial part of the wavefunction we can simplify this to just the radial part of the wavefunction:

##### eq. 2
$ \psi (Radial) (r) = R(r) $  

Even though we have been able to simplify this to just being the radial part, in terms of n and l, the equation is still quite complicated. (See below)

$R_{nl}(r) = -\left(\frac{(n-l-1)!}{2n(n+1!)^3}\right)^\frac{1}{2} \left( \frac{2}{na_0}\right)^{l+\frac{3}{2}} r^l e^{\frac{-r}{na_0}}L_{n+l}^{2l+1}\frac{2r}{na_0}$

Where, $L_{n+l}^{2l+1} $ are polynomials called associated Laguerre polynomials.

Rather than trying to work out each of these in the general case as above, in this worksheet we are going to make life a little easier and just use the solutions to this equation for some orbitals. These are summarized in the table below.


| Radial Part $R(r)$ |
|:---------------------------------------------------------------------------------------------------------------------------------------:|
| $R(1s) = 2 (\frac{Z}{\alpha _0})^\frac{3}{2}e^\frac{-\sigma}{2} $                                                                                       | 
| $R(2s) = \frac{1}{2\sqrt{2}} (\frac{Z}{\alpha _0})^\frac{3}{2}(2-\sigma)e^\frac{-\sigma}{2} $                                                                       |
| $R(3s) = \frac{1}{9\sqrt{3}} (\frac{Z}{\alpha _0})^\frac{3}{2}(6-6\sigma + \sigma^2)e^\frac{-\sigma}{2} $   | 
|  $R(2p) = \frac{1}{\sqrt{24}} (\frac{Z}{\alpha _0})^\frac{3}{2}e^\frac{-\sigma}{2}$  | 
|  $R(3p) = \frac{1}{9\sqrt{6}} (\frac{Z}{\alpha _0})^\frac{3}{2}(4-\sigma)\sigma)e^\frac{-\sigma}{2}$  | 
|  $R(3d) = \frac{1}{9\sqrt{30}} (\frac{Z}{\alpha _0})^\frac{3}{2}\sigma^2e^\frac{-\sigma}{2}$  | 

Where, $ \sigma = \frac{2Zr}{na_0}$, na_0 is the Bhor radius 52.9177 pm, Z is the atomic number  (1 for hydrogen), and r is the distance away from the nucleus.

Great, now that we know sigma is defined properly. We really want to plot the radial part of the wavefunction not sigma. So in the block below define a new funtion Radial_1s(r) this will be the radial part of the 1s wavefunction! It's in the chart above. 
<br> $R(1s) = 2 (\frac{Z}{\alpha _0})^\frac{3}{2}e^\frac{-\sigma}{2} $   
<br> a couple of hints here, <br> 1) because we defined sigma already you can use it in the function directly as sigma(r). <br> 2) To code in the exponential we need to use numpy, to call it write np.exp() and put the stuff you want to have in the exponential in side the brackets.

```python
def Radial_1s(r):
    return 2*(Z/ao)*np.exp(-sigma(r)/2);
```

One way of interpeting the wavefunction is to look at the square of its value, normalizing it and treating this as the probabiltiy density. 

> $ Probabiltiy density = 4 \pi r^2 * R(r)^2 $

In the code block below define a new function for the probabiltiy denisty. The python way of squaring a function is to write two multiplation symbols followed by the number 2. ex. 3 ** 2 = 9

```python
def probabilitydensity1s(r):
    return 4*np.pi*r**2 * Radial_1s(r) ** 2
```

Define functions for the rest of the of the orbitals in the chart above (2s, 2p, 3s, 3p and 3d oribtals), we'll go through the 2p orbital as an example.

We start by defining and naming the function:
> def twos(r): 

In the next line we are going to enter a new term x, this is going to the value of n for your wavefunction. Since we are doing the 2p orbital x=2 (we do this because sigma is a function of n but we have already defined it).

```python
def twop(r):
    x=2
    return (1/np.sqrt(24))*(Z/ao)**(3/2)*(sigma(r)/x)*np.exp(-(sigma(r)/x)/2);
```

```python
def threes(r):
    x=3
    return (1/(9*np.sqrt(3)))*((Z/ao)**(3/2))*(6-(6*(sigma(r)/x))+((sigma(r)/x)**2))*(np.exp(-(sigma(r)/x)/2));
```

```python
def threep(r):
    x=3
    return (1/(9*np.sqrt(6)))*((Z/ao)**(3/2))*(4-(sigma(r)/x))*(sigma(r)/x)*(np.exp(-(sigma(r)/x)/2));
```

```python
def threed(r):
    x=3
    return (1/(9*np.sqrt(30)))*((Z/ao)**(3/2))*(sigma(r)/x)**2*np.exp(-(sigma(r)/x)/2);
```

## 5. Define the probability Functions.

Now define the probability density function of each of the orbitals in the code blocks below, don't forget to normalize them all with $4 \pi r^2$

```python
def probabilitydensity3p(r):
    return (threep(r) ** 2)*r*r*4*np.pi
```

```python
def probabilitydensity2p(r):
    return  r*r*twop(r)*twop(r)*4*np.pi
```

```python
def probabilitydensity3d(r):
    return (threed(r) ** 2)*(r*r)*4*np.pi
```

```python
def probabilitydensity2s(r):
    return   r*r*twos(r)*twos(r)*4*np.pi
```

```python
def probabilitydensity3s(r):
    return   (threes(r) ** 2)*(r*r)*4*np.pi
```
